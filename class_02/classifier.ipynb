# Prediction(regression) vs Classification

# what they have in common
they both use historical data
they both include features and target
SUPERVISED LEARNING --> model learns from data with known features (labeled data)

# what is the difference
prediction estimates and returns a specific value + error 
classificaton returns 1 or 0 depending on the label 
different models (linear regression --> numeric values only (suitable, the baseline for prediction) (Logistic regression, baseline for classification)
dirrerent evaluation metrics
Prediction(regression)
metrics: mean absolute error, root mean square error 
Classification
metrics: confusion matrix (TF, TN, FP, FN)
accurancy

```python
# import libraries or classes from libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv("diabetes.csv")
data.head()
```

```python
data.info()
```

```python
# descriptive analytics of the dataset
# EDA
data.describe()
```

```python
# we have no missing values but we have values that do not make sense, for example 0 BMI
# we are going to treat zeros as missing values

(data[['Glucose','BloodPressure','SkinThickness','BMI']] == 0).sum()
```

```python
# replace O with NaN
columns = ['Glucose', 'BloodPressure', 'SkinThickness', 'BMI']
data[columns] = data[columns].replace(0, np.nan)
data.isnull().sum()
```

```python
# replace the NaN with the median (interpolation)
data.fillna(data.median(), inplace=True)
data.isnull().sum()
```

```python
data.head()
```

```python
# define features and target for classification
# X (features) and y (target)
X = data.drop('Outcome', axis=1) ## is a matrix
y = data['Outcome'] ## is a vector
```

```python
# split the data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# apply the model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
```
